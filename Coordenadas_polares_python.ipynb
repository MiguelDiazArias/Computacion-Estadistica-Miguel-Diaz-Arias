2. Graficar un cardioide en coordenadas polares en Python

```python
import pandas as pd
import numpy as np
def prog(r, n, a1):
  an = a1 + r*(n-1)
  seq = np.arange(start=a1, stop=an, step=r)
  return seq

prog(r=7, n=20, a1=15)

```

Primera matriz de datos

```python
np.random.seed(123)

df1 = pd.DataFrame({
    'de': np.sort(np.random.normal(loc = 4, scale = 1, size=96)),
    'dl': np.sort(np.random.normal(loc=4.5, scale=1.2, size=96)),
    'ddd': np.repeat(prog(r=7, n=25, a1=15), 4)
})

df1['localidad'] = np.repeat(['l1','l2']*24, 2)
df1.head()
```

Convertir "de" y "dl" a coordenadas polares

```python
np.random.seed(123)
x=df1['de']
y=df1['dl']

r=np.sqrt(x**2 + y**2) ##Radio polar
tr=np.arctan2(y,x) ## Angulo polar en raides
import numpy
tc=numpy.degrees(tr)
print(f'La distancia desde el origen "r" {r}')
print(f'El angulo formado "t" es {tc}')

```

Graficar una cardioide en coordenadas polares en Python

```python
from matplotlib import pyplot as plt

t = np.linspace(0, 2*np.pi, 1000)

r = 5 - 5 * np.sin(t)

plt.polar(t, r, 'r')

plt.show()
```
